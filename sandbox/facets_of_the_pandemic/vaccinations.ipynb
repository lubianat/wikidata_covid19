```python
from wikidata2df import wikidata2df
```

```python
query_for_vaccinations = """
SELECT 
  DISTINCT 
  # Remove ?item and ?title to generate table
  # in Quickstatements V1 syntax
  ?item ?itemLabel

WHERE {

  ?item wdt:P31 wd:Q104549533.
  MINUS{?item wdt:P1269 ?of.}
  
  ?item rdfs:label ?label.
  
  FILTER(lang(?label)="en")
  FILTER(CONTAINS(lcase(?label), "covid"))
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }

}
"""

vaccinations = wikidata2df(query_for_vaccinations)
```

```python
query_for_countries = """
SELECT DISTINCT ?item ?itemLabel ?country ?countryLabel WHERE { 
  ?item p:P31 [
              ps:P31 wd:Q3241045;
              pq:P642 wd:Q84263196;
              pq:P3005 ?country
              ].
  values ?type_of_country {wd:Q3624078 wd:Q6256}.
  ?country wdt:P31 ?type_of_country.

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". } 
}
ORDER BY ?country ?dates
"""

countries = wikidata2df(query_for_countries)
```
