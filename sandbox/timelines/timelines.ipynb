```python
from wikidata2df import wikidata2df
```

```python
query_for_timelines = """
SELECT 
  DISTINCT 
 ?item ?itemLabel
?of ?ofLabel
WHERE {
 ?item p:P31 [ps:P31 wd:Q18340550;
                     pq:P642 wd:Q81068910].  
  
 ?item rdfs:label ?label
       
 OPTIONAL {?item wdt:P1269 ?of}
    SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
"""

timelines = wikidata2df(query_for_timelines)
```

```python
query_for_countries = """
SELECT DISTINCT ?item ?itemLabel ?country ?countryLabel WHERE { 
  ?item p:P31 [
              ps:P31 wd:Q3241045;
              pq:P642 wd:Q84263196;
              pq:P3005 ?country
              ].
  values ?type_of_country {wd:Q3624078 wd:Q6256}.
  ?country wdt:P31 ?type_of_country.

  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". } 
}
ORDER BY ?country ?dates
"""

countries = wikidata2df(query_for_countries)
```

```python
for index, row in timelines.iterrows():
    
    timeline_qid = row["item"]
    timeline_label = row["itemLabel"]
    
    for index2, row2 in countries.iterrows():
        
        if row2["itemLabel"].lower() in timeline_label.lower():
            pandemic_qid = row2["item"]
            print(timeline_qid + "|P1269|" + pandemic_qid + "|S887|Q84423633" )
    
```
