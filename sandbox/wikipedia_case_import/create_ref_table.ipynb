```python
import pandas as pd
from wikidata2df import wikidata2df

query = """
SELECT ?item ?itemLabel ?countryLabel
WHERE 
{
  ?item p:P31 ?statement.
       ?statement ps:P31 wd:Q3241045.
       ?statement pq:P642 wd:Q84263196.
       ?statement pq:P3005 ?country.
  
  VALUES ?countryTypes {wd:Q6256 wd:Q3624078 wd:Q1048835}.
  
  ?country wdt:P31 ?countryTypes.
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}
"""

local_outbreak_items = wikidata2df(query)
```
