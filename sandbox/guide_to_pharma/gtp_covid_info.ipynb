### Guide to Pharmacology - COVID19 release

I believe more information from guide to Pharmacology can be integrated, but the least that can be done is integrating their IDs in Wikidata.

```python
import pandas as pd
```

### Ligands

```python
covid_ligands = pd.read_csv("./gtopdb_ligands_reconciled.csv")
covid_ligands
```

```python
has_qid = covid_ligands.query("qid.notnull() & `Ligand ID`.notnull()")
has_qid
```

```python
for index,row in has_qid.iterrows():
    ligand_id = str(int(row['Ligand ID']))
    print(
    row['qid'] + "|P595|" + '"' + ligand_id + '"' + "|S854|" + '"' + 
        'https://www.guidetopharmacology.org/coronavirus.jsp' + '"' +
        "|S813|" + "+2020-04-22T00:00:00Z/11"
    )
```
