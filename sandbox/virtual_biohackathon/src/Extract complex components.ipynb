```python
import pandas as pd

complexes_data = pd.read_csv("../data/sars-cov-2.tsv", sep = "\t")
complexes_data.head(2)
```

```python
# Okay, now let's create a ID to Wikidata reference object.

# From https://covid-19.uniprot.org/uniprotkb/P0DTD1 we imported infos to OpenRefine

# We created the items for the proteins derived of P0DTD1-PRO_0000449628.

open_refine_reference = pd.read_csv("../intermediate/openrefine_replicase_frags.tsv", sep="\t")
open_refine_reference["code_as_in_complexes_data"] = open_refine_reference["PDB ID"] + "-" + open_refine_reference["ID"] 
open_refine_reference.head(2)
```

```python
# We also need to get  the ids from the complexes in Wikidata, of course. Once again, OpenRefine is the way to go.
open_refine_complexes = pd.read_csv("../intermediate/openrefine_complexes.tsv", sep="\t")
open_refine_complexes.head(2)
```

```python
# Cool, but some repetion of subunits are not identified in the (stoichiometry) part. 
# I guess I could write code to clean that up, but for now I will do it manually. 

open_refine_complexes_manually_edited = pd.read_csv("../intermediate/openrefine_complexes_manual_edit.tsv", sep="\t")
open_refine_complexes_manually_edited.head()
```

```python
# Convert to quickstatements only the complexes that are derived of P0DTD1: 


! touch ../p0dtd1_quickstatements.txt
! :> ../p0dtd1_quickstatements.txt
for index, row in open_refine_complexes_manually_edited.iterrows():
    complex_composition = row["Identifiers (and stoichiometry) of molecules in complex"]
    complex_qid = row["complex_qid"]
    cpx_id = row["#Complex ac"]
    source = '|S248|Q47196990|S7718|"' + cpx_id + '"|S813|+2020-04-10T00:00:00Z/11|S854|"ftp://ftp.ebi.ac.uk/pub/databases/intact/complex/current/complextab"|S7793|"sars-cov-2.tsv"'
  
    if "P0DTD1" in complex_composition:
        print("Adding QuickStatements for " + cpx_id) 
        ids_in_this_complex = complex_composition.split("|")
        
        for id_and_amount in ids_in_this_complex:
            
            id_and_amount = id_and_amount.replace(")", "")
            id_and_amount_tuple = id_and_amount.split("(")
            subunit_qid = open_refine_reference["qid_updated"][open_refine_reference["code_as_in_complexes_data"] ==id_and_amount_tuple[0]]
            amount = id_and_amount_tuple[1]
            if amount == 0:
            # 0 means unknown    
                quickstatements_code = complex_qid + "|P527|" + subunit_qid + source
            else:
                quickstatements_code = complex_qid + "|P527|" + subunit_qid + "|P1114|" + amount + source
            with open("../p0dtd1_quickstatements.txt", "a") as myfile:
                myfile.write(quickstatements_code.values[0] + "\n")
            
            
```

```python
# Lets now see which ones are not from P0DT1 and update those. 
for index, row in open_refine_complexes_manually_edited.iterrows():
    cpx_id = row["#Complex ac"]
    complex_composition = row["Identifiers (and stoichiometry) of molecules in complex"]
    if "P0DTD1" not in complex_composition:
        print(cpx_id)
        print(complex_composition)
```

```python
# Okay, so now we need the reference proteins for these complexes.
reference_id_to_wikidata = {
"Q9BYF1":"Q301630",
"Q695T7":"Q21124449",
"P0DTC9":"Q87917584",
"CPX-5682":"Q90012262",
"P0DTC2-PRO_0000449648":"mock",
"P0DTC2-PRO_0000449647":"Q90081967"
    
}
```

```python
# Convert to quickstatements only the complexes that are not derived of P0DTD1: 


! touch ../non_p0dtd1_quickstatements.txt
! :> ../non_p0dtd1_quickstatements.txt

for index, row in open_refine_complexes_manually_edited.iterrows():
    complex_composition = row["Identifiers (and stoichiometry) of molecules in complex"]
    complex_qid = row["complex_qid"]
    cpx_id = row["#Complex ac"]
    source = '|S248|Q47196990|S7718|"' + cpx_id + '"|S813|+2020-04-10T00:00:00Z/11|S854|"ftp://ftp.ebi.ac.uk/pub/databases/intact/complex/current/complextab"|S7793|"sars-cov-2.tsv"'
  
    if "P0DTD1" not in complex_composition:
        print("Adding QuickStatements for " + cpx_id) 
        ids_in_this_complex = complex_composition.split("|")
        
        for id_and_amount in ids_in_this_complex:
            id_and_amount = id_and_amount.replace(")", "")
            id_and_amount_tuple = id_and_amount.split("(")
            subunit_qid = reference_id_to_wikidata[id_and_amount_tuple[0]]
            amount = id_and_amount_tuple[1]
            if amount == 0:
            # 0 means unknown    
                quickstatements_code = complex_qid + "|P527|" + subunit_qid + source
            else:
                quickstatements_code = complex_qid + "|P527|" + subunit_qid + "|P1114|" + amount + source
            with open("../non_p0dtd1_quickstatements.txt", "a") as myfile:
                myfile.write(quickstatements_code + "\n")
```
